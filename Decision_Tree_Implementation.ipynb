"chas" has least significance on medv so remove it 

```python
df = df.drop(["chas"], axis=1)

df.head()
```

```python
df.describe()
```

```python
def adjust(val, irq, q1, q3):
    high = q3 + irq * 1.5
    low = q1 - irq * 1.5

    if(val > high):
        return high
    elif (val < low):
        return low
    else:
        return val

def remove_outlier(feature):
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)

    irq = q3 - q1

    df[feature] = [adjust(i, irq, q1, q3) for i in df[feature]]

remove_outlier("crim")
remove_outlier("zn")
remove_outlier("rm")
remove_outlier("dis")
remove_outlier("b")
remove_outlier("ptratio")
remove_outlier("lstat")
```

## Spliting the Data

```python
x = df.drop(["medv"], axis=1).values
y = df["medv"].values

x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

## Preprocessing

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_test
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

y_pred
```

```python
y_test
```

## Error Calculation

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

## Visualization

```python
plt.plot(x_test, y_test,"+", color = "green", label="Actual Data")
plt.plot(x_test, y_pred,".", color = "red", label="Predicted Data")
plt.title("Predicted VS Actual Data")
plt.show()
```
