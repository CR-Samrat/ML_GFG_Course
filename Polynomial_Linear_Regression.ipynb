```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./DataSets/Position_Salaries.csv')

df.head()
```

## Spliting the data

```python
x = np.array(df.iloc[:,1:-1])
y= np.array(df.iloc[:,-1])

x
```

```python
y
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)
```

```python
plt.scatter(x, y, label="Actual Data")
plt.plot(x, model.predict(x), color="red", label="Prediction")
plt.legend()
plt.show()
```

## Polynomial Linear Regression

**Degree 2**

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)

x_poly = poly_reg.fit_transform(x)

x_poly
```

```python
model_2 = LinearRegression()

model_2.fit(x_poly, y)

plt.scatter(x, y, label="Actual Data")
plt.plot(x, model_2.predict(x_poly), color="red", label="Prediction")
plt.title("Polynomial Regression with degree 2")
plt.legend()
plt.show()
```

**Degree 3**

```python
poly_reg = PolynomialFeatures(degree=3)

x_poly = poly_reg.fit_transform(x)

x_poly
```

```python
model_3 = LinearRegression()

model_3.fit(x_poly, y)

plt.scatter(x, y, label="Actual Data")
plt.plot(x, model_3.predict(x_poly), color="red", label="Prediction")
plt.title("Polynomial Regression with degree 3")
plt.legend()
plt.show()
```
