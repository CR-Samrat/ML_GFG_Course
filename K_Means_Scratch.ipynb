## The code will not work if the random initialization of centroid is not well

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Creating the dataset

```python
from sklearn.datasets import make_blobs

x,y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=3)

x
```

```python
y
```

## Visualizing the dataset

```python
plt.grid(True)
plt.scatter(x[:,0], x[:,1], c = y)
plt.title("Input Data Points")
plt.show()
```

## Initialize K Random Centroids

```python
# As we can see value is between -10 to +10
def gen_random():
    return 10*(2*np.random.random(2)-1)

print(gen_random())
```

```python
k = 5
clusters = {}
colors = ["red","green","blue","yellow","cyan"]

for idx in range(k):
    center = gen_random()
    cluster = {
        'points': [],
        'center': center,
        'color': colors[idx]
    }

    clusters[idx] = cluster

clusters
```

## Visualize the initial centroids

```python
for i in clusters:
    print(clusters[i])
```

```python
plt.grid(True)
plt.scatter(x[:,0], x[:,1], c = y)
for i in clusters:
    plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1], marker='^', s=70, c="red")
plt.title("Initial Cluster Centroids")
plt.show()
```

## Assigning each data points to it's nearest cluster

```python
def distance(x, y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
#Testing distance function

x_1 = np.array([1,2])
x_2 = np.array([3,4])

distance(x_1,x_2)
```

```python
def assign_cluster():
    for i in x:
        cur_x = i
        dist = []
        
        for j in clusters:
            dist.append(distance(cur_x,clusters[j]["center"]))
        
        # argmin will return the index of the smallest element
        cur_cluster = np.argmin(dist)
        clusters[cur_cluster]['points'].append(cur_x)

        dist = []

assign_cluster()
```

```python
for i in clusters:
    print(clusters[i]['points'])
```

## Visualizeing the new clusters 

```python
def plot_clusters():
    plt.grid(True)
    for i in clusters:
        pts = np.array(clusters[i]['points'])
        if(len(pts) == 0):
            continue
        plt.scatter(pts[:,0],pts[:,1], c=clusters[i]['color'])
        plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1], marker='^', s=70, c="black")
    plt.title("After assigning the data points to Clusters")
    plt.show()
```

```python
plot_clusters()
```

## Updating the centroids 

```python
def update_centroid():
    for i in clusters:
        pts = np.array(clusters[i]["points"])
        
        new_centroid = pts.mean(axis=0)
        clusters[i]["center"] = new_centroid
        clusters[i]["points"] = []
        print("Centroids",(i+1)," :",clusters[i]['center'])

update_centroid()
```

## Run all the functions until centroid become unchangeable

```python
for _ in range(10):
    assign_cluster()
    plot_clusters()
    update_centroid()
```
